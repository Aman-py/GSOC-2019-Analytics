```python
d={'org_name':[],'category':[],'tech':[],'topics':[]}
i = 0
for x in data_ids[153:]:
    flag = True
    while flag:
        try:
            browse = webdriver.PhantomJS('/home/aman-py/phantomjs-2.1.1-linux-x86_64/bin/phantomjs')
            browse.get('https://summerofcode.withgoogle.com/organizations/?sp-page=5#'+x)
            source = browse.page_source
            soup = bs4.BeautifulSoup(source,'lxml')
            org_name=soup.find('h2',{'class':'organization-card__title md-display-2'}).text
            print(i)
            print(org_name)
            zp=soup.find_all('li',{'class':'organization__tag organization__tag--technology'})
            tech = [i.text for i in zp]
            print(tech)
            category = soup.find('li',{'class':'organization__tag organization__tag--category'}).text
            print(category)
            tp=soup.find_all('li',{'class':'organization__tag organization__tag--topic'})
            topics = [j.text for j in tp]
            print(topics)
            d['org_name'].append(org_name)
            d['category'].append(category)
            d['tech'].append(tech)
            d['topics'].append(topics)
            i+=1
            flag = False
        except AttributeError:
            print('Attribute Error Occured')
        
    
    #print(x)
    #print(org_name)
    #print(tech)
    #print(category)
    #print(topics)
```
